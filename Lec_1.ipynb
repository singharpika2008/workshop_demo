### Comprehensive Pandas Demonstration
Using the California Housing dataset to explore library features.

````python
# 1. Loading the dataset
housing_df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# 2. Basic Inspection
print("Dataset Shape:", housing_df.shape)
housing_df.head()
````

````python
# 3. Data Info and Descriptive Statistics
# Provides data types and non-null counts
print(housing_df.info())

# Provides summary statistics for numerical columns
display(housing_df.describe())
````

````python
# 4. Data Selection and Filtering
# Selecting specific columns and rows where median_house_value is high
high_value_houses = housing_df[housing_df['median_house_value'] > 400000][['longitude', 'latitude', 'median_house_value']]

print("High Value Houses Preview:")
display(high_value_houses)
````

````python
# 5. Feature Engineering (Creating new columns)
# Calculate average rooms per household
housing_df['rooms_per_household'] = housing_df['total_rooms'] / housing_df['households']

display(housing_df[['total_rooms', 'households', 'rooms_per_household']].head())
````

````python
# 6. Grouping and Aggregation
# Grouping by housing_median_age to see average value
age_group_summary = housing_df.groupby('housing_median_age')['median_house_value'].mean().reset_index()

print("Average House Value by Age Group:")
display(age_group_summary.head())
````

### 7. Data Cleaning
This section demonstrates how to handle null values, duplicates, and data types.

````python
# Check for missing values in each column
print("Missing values per column:")
print(housing_df.isnull().sum())

# Check for duplicate rows
print("\nNumber of duplicate rows:", housing_df.duplicated().sum())

# Verify data types to ensure they are appropriate for analysis
print("\nData Types:")
print(housing_df.dtypes)
````

If there were missing values, we could use `housing_df.fillna(value)` to replace them or `housing_df.dropna()` to remove those rows. Similarly, `housing_df.drop_duplicates()` would be used to remove redundant data.
